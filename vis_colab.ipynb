```python
from save_and_load import *
import matplotlib.pyplot as plt

path = "/content/drive/MyDrive/dqn_t1/"


loss_mean = obj_load(path+"loss")
q_mean = obj_load(path+"q")
rewards = obj_load(path+"rewards")

import IPython
IPython.display.clear_output()

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].plot(loss_mean[25:])
axs[0].set_title('Loss Mean')

axs[1].plot(q_mean[25:])
axs[1].set_title('Q Mean')

rewards_sum = [0]
for r in rewards:
    rewards_sum.append(rewards_sum[-1]+r)

#axs[2].plot(rewards)
axs[2].plot(rewards_sum)
axs[2].axhline(0, color='r', linestyle='--')
axs[2].set_title('Rewards')

plt.tight_layout()
plt.show()
```
